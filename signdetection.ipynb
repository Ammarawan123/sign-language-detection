```python
from ultralytics import YOLO

model=YOLO('runs/detect/train/weights/best.pt')

metrics=model.val()

print("mAP:",metrics.box.map)
print("mAP50:",metrics.box.map50)
print("mAP75:",metrics.box.map75)
print("Per-class mAP:",metrics.box.maps)
```
